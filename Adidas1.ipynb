```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df1=pd.read_excel('C:/Users/y20cs99/Downloads/Adidas_USSales_Datasets.xlsx',header=4)
df1
df=df1.dropna(axis=1,how="all")
df
```

```python
#1.	List all the retailers with retailer id.
df[['Retailer','Retailer ID']].drop_duplicates()

```

```python
#2.	List all the retailers in every region.
dg=df.groupby('Region')['Retailer'].unique()
dg
#for n,m in dg:
#print(n,'***',m,sep='\n')
```

```python
#3.	List the retailers in every city of a state.
dg=df.groupby(['City','State'])['Retailer'].unique()
dg
#for n,m in dg:
#print(n,'******',m,sep='\n')
```

```python
#4.	List the products sold by the retailer in every city.
dg=df.groupby(['Retailer','City'])['Product'].unique()
dg
```

```python
#5.	Find the total sales of every retailer.
dg=df.groupby('Retailer')['Total Sales'].sum()
dg
#for n,m in dg:
#print(n,np.sum(m['Total Sales']),sep=' ')
```

```python
#6.	Find the total sales of the retailers in every city along with profit.
dg=df.groupby(['Retailer','City'])[['Total Sales','Operating Profit']].sum()
#for n,m in dg:
#print(n,np.sum(m['Total Sales']),np.sum(m['Operating Profit']))
dg
```

```python
#7.	Find the total sales &profit of each product sold by the retailer.
dg=df.groupby('Product')[['Total Sales','Operating Profit']].sum()
dg
```

```python
#8.	Find the units sold, total sales &profit of the products sold between the dates 1/1/2020 and 4/15/2020.
s=datetime.strptime('1/1/2020','%m/%d/%Y')
e=datetime.strptime('4/15/2020','%m/%d/%Y')
df[df['Invoice Date'].between(s,e)][['Units Sold','Total Sales','Operating Profit']]
```

```python
#9.	Find the no of units sold of each product by each retailer in every city.
dg=df.groupby(by=['Product','Retailer','City'])['Units Sold'].sum()
dg
```

```python
#12.	Find the total sales &profits of all products in every month.
dg=df.groupby([df['Invoice Date'].dt.month,'Product'])[['Total Sales','Operating Profit']].sum()
dg
```

```python
#13.	Find the total sales &profit of the products in different sales methods in each city.
dg=df.groupby(['Product','City','Sales Method'])[['Total Sales','Operating Profit']].sum()
dg
```
