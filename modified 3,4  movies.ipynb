```python
import numpy as np
import pandas as pd
from dateutil.parser import parse
```

```python
df=pd.read_csv(r"C:\Users\Sravani\Downloads\movies_1_.csv")
df
```

```python
a=df[df['Release Date'].str.contains('TBD')]
l=a.index.to_list()
print(l)
df.drop(l,axis=0,inplace=True)
```

```python
for i in range(len(df)):
    df.iloc[i,5]=parse(df.iloc[i,5])
```

```python
df['Release Date']=pd.to_datetime(df['Release Date'])
print(df['Release Date'])
df['year']=df['Release Date'].dt.year
df['month']=df['Release Date'].dt.month
df['day']=df['Release Date'].dt.day
df
```

```python
#Find the movies released world-wide and find out the revenue received world-wide other than US with their ratings.
l=df[df['US Gross']=='0'].index
df4=df.drop(l)
df4['US Gross']=df4['US Gross'].replace('Unknown',np.nan)
df4['Worldwide Gross']=df4['Worldwide Gross'].replace('Unknown',np.nan)
m=df4[df4['Worldwide Gross'].isna()].index
df4.drop(m,inplace=True)
df4['Worldwide Gross']=pd.to_numeric(df4['Worldwide Gross'])
df4['US Gross']=pd.to_numeric(df4['US Gross'])
df4['revenue']=df4['Worldwide Gross']-df4['US Gross']
print(df4[['Title','revenue','IMDB Rating']])
```

```python
#). Find the movies with loss & profit released in each year with genre and ratings.
df['US Gross']=df['US Gross'].replace('Unknown',np.nan)
df['Worldwide Gross']=df['Worldwide Gross'].replace('Unknown',np.nan)
df['Worldwide Gross']=pd.to_numeric(df['Worldwide Gross'])
df['US Gross']=pd.to_numeric(df['US Gross'])
df['Production Budget']=pd.to_numeric(df['Production Budget'])
df['loss/profit']=df['Production Budget']-(df['Worldwide Gross']+df['US Gross'])
df
```

```python
dg=df.groupby('year')
for i,j in dg:
    print(i,j[['Title','loss/profit','Major Genre','IMDB Rating']],sep="\n")
```
